```python
import numpy as np
import coilpy as cp
from matplotlib import pyplot as plt
```

```python
equi = cp.vmec.VMECout("wout_zot80-m.00000.nc")
```

```python
surfs = equi.surface
nsurf = len(surfs)
a_vessel = 0.072
R_vessel = 0.3048
```

# Surfaces

```python
equi.wout
```

```python
# %matplotlib ipympl
%matplotlib inline
n = 50
# pi/(2*)
us = np.pi/2*np.linspace(0,1,5).reshape((-1,1))*np.linspace(1,1,n)
print(us)
v = np.linspace(0,2*np.pi,n)
print(v)
# points = [surfs[i].rz(v, u) for i in range(nsurf)]
r = 0.3048 + 0.072 * np.cos(v)
z = 0.072 * np.sin(v)

fig, ax = plt.subplots(1, 5)
# fig, ax = plt.subplots(2, 5)
fig.set_figwidth(50)
fig.set_figheight(20)

for j in range(5):
    u = us[j,:]
    points = [surfs[i].rz(v, u) for i in range(nsurf)]
    # points1 = [muse.rzpl(v, u, i)[:2] for i in range(nsurf)]
    for i in range(0,30+1):
        ax[j].plot(points[i][0],points[i][1])
        # ax[1, j].plot(points1[i][0],points1[i][1])

    ax[j].plot(r,z,'--')    
    ax[j].set_aspect(1)
    # ax[1,j].plot(r,z,'--')    
    # ax[1,j].set_aspect(1)
    # ax[j].plot(muse.RZ_axis(u[0])[0], muse.RZ_axis(u[0])[1] , '*')

plt.show()
```
